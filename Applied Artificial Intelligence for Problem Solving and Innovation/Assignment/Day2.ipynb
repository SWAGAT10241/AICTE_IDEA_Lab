```python
import pandas as pd

data = {
    'Student':['Amit','Ravi','Sita','Priya','Rahul','Neha','Arjun','Pooja','Karan','Meera'],
    'Hours_Studied': [2, 4, 6, 8, 10, 5, 7, 3, 9, 6],
    'Attendance(%)': [75, 80, 85, 90, 95, 82, 88, 78, 92, 86],
    'Assignments_Submitted': [4, 5, 6, 7, 8, 5, 6, 4, 7, 6],
    'Previous_Score': [32, 45, 58, 72, 88, 52, 68, 40, 80, 62],
    'Marks': [35,50,65,80,95,60,75,45,88,70],
}

pf = pd.DataFrame(data)

pf
```

2 . Calculate:
- Mean,median and standard deviation of Marks
- Mean attendance
- Maximum and minimum Previous_Score

```python
print(f"Mean : {pf['Marks'].mean()}")
print(f"Median : {pf['Marks'].median()}")
print(f"Standard deviation : {pf['Marks'].std()}")
```

```python
print(f"Maximum Previous Score : {pf['Previous_Score'].max()}")
print(f"Minimum Previous Score : {pf['Previous_Score'].min()}")
```

3 . Find

- Student with the second-higest Marks

```python
print(f"Student with the higest Marks : {pf['Marks'].max()}")
print(f"Student with the second-higest Marks : {pf['Marks'].nlargest(2).iloc[-1]}")
print(f"Difference between their scores : {pf['Marks'].max() - pf['Marks'].nlargest(2).iloc[-1]}")
```

5 . Display all students who satify both conditions:
- Attendance >= 85 %
- Marks >= 70 <br>
Also calculate their average Marks.

```python
students = pf[(pf['Marks'] >= 70) & (pf['Attendance(%)'] >= 85)]
print(students)

print(f"\nAverage Marks of these students: {students['Marks'].mean():.2f}")
```

6. Create a new feature:
Performance_Index = (Marks + Previous_Score)/2
Find:
- Student with the highest Performance_Index
- Average Performance_Index of the class

```python
Performance_Index = ((pf['Marks']) + (pf['Previous_Score']))/2
print(f'Highest Performace Index : {Performance_Index.max()}')
print(f'Average Performace Index : {Performance_Index.mean()}')
```
