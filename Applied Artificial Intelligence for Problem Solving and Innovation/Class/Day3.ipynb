## Feature Selection and Correlation

We can use selective features to improve model performance and reduce complexity. This can be accomplished using the correlation function, which outputs a value between -1 and 1:

*   **1**: Denotes a perfect positive correlation (as one variable increases, the other increases proportionally).
*   **0**: Denotes no linear correlation between the variables.
*   **-1**: Denotes a perfect negative correlation (as one variable increases, the other decreases proportionally).

Correlation helps in understanding the strength and direction of a linear relationship between two numerical variables. A higher absolute correlation value (closer to 1 or -1) indicates a stronger relationship, which can be useful for selecting relevant features for modeling.

## Key Theories in Data Analysis and Machine Learning:

*   **Descriptive Statistics**: Summarizes and describes the main features of a collection of information. It includes measures like mean, median, mode, standard deviation, and variance, which help in understanding the central tendency, spread, and shape of data distributions.

*   **Inferential Statistics**: Uses a random sample of data taken from a population to describe and make inferences about the whole population. This involves hypothesis testing, confidence intervals, and regression analysis to draw conclusions and make predictions.

*   **Linear Regression**: A fundamental statistical model that examines the linear relationship between a dependent variable and one or more independent variables. It's often used for predicting continuous outcomes.

*   **Data Preprocessing**: Encompasses techniques for cleaning, transforming, and integrating data to make it suitable for analysis or machine learning models. This includes handling missing values (e.g., imputation, deletion), normalizing/scaling data, and encoding categorical variables.

*   **Data Visualization**: The graphical representation of information and data. By using visual elements like charts, graphs, and maps, data visualization tools provide an accessible way to see and understand trends, outliers, and patterns in data.

```python
import pandas as pd
data = {
    'Student':['Amit','Ravi','Sita','Priya','Rahul','Neha','Arjun','Pooja'],
    'Hours_Studied': [2, 4, 6, 8, 10, 5, 7, 3],
    'Attendance' : [75,80,85,90,95,82,88,78],
    'Marks': [35,50,65,80,95,60,75,45]
}

df = pd.DataFrame(data)
df
```

## **Histogram**


A histogram is a graphical representation of the distribution of numerical data. It is an estimate of the probability distribution of a continuous variable and was first introduced by Karl Pearson.

In the code:

*   `import matplotlib.pyplot as plt` and `import seaborn as sns`: These lines import the necessary libraries for plotting.
*   `plt.figure(figsize=(8, 6))`: This sets the size of the plot.
*   `sns.histplot(df['Marks'], kde=True, bins=5)`: This is the core function call from `seaborn`. It creates a histogram for the 'Marks' column of your DataFrame `df`.
    *   `kde=True`: This overlays a Kernel Density Estimate (KDE) curve, which is a smoothed representation of the data's distribution.
    *   `bins=5`: This parameter divides the range of 'Marks' into 5 intervals (bins) and counts how many marks fall into each bin. You can adjust this number to see more or less detail in the distribution.
*   `plt.title('Distribution of Marks')`, `plt.xlabel('Marks')`, `plt.ylabel('Frequency')`: These lines add a title and labels to the x and y axes for clarity.
*   `plt.grid(True, linestyle='--', alpha=0.7)`: This adds a grid to the plot, making it easier to read.
*   `plt.show()`: This displays the generated plot.

```python
import matplotlib.pyplot as plt

plt.hist(df['Marks'], bins=5, edgecolor='black')
plt.title('Distribution of Marks')
plt.xlabel('Marks')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

## **Box Plot**

A **boxplot** (also known as a box-and-whisker plot) is a standardized way of displaying the distribution of data based on a five-number summary: minimum, first quartile (Q1), median (Q2), third quartile (Q3), and maximum.

In the box plot for 'Marks' that was generated:

*   **The Box**: The rectangular box itself represents the **interquartile range (IQR)**, which spans from the first quartile (Q1) to the third quartile (Q3). This box contains the middle 50% of the data.
    *   The **bottom edge** of the box is the **First Quartile (Q1)**, meaning 25% of the data falls below this value.
    *   The **line inside the box** is the **Median (Q2)**, which is the middle value of the dataset, dividing it into two equal halves (50% of data below, 50% above).
    *   The **top edge** of the box is the **Third Quartile (Q3)**, meaning 75% of the data falls below this value.

*   **The Whiskers**: These lines extend from the box to the minimum and maximum values within a certain range (usually 1.5 times the IQR from Q1 and Q3). They indicate the variability outside the middle 50% of the data.

*   **Outliers**: Any data points that fall outside the whiskers are considered potential outliers and are often plotted as individual points (dots or asterisks). This plot does not show any explicit outliers.

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
sns.boxplot(y=df['Marks'])
plt.title('Boxplot of Marks')
plt.ylabel('Marks')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

## **Correlation Matrix**

A **Correlation Matrix** is a table showing correlation coefficients between variables. Each cell in the table shows the correlation between two variables. The value ranges from -1 to 1:

*   **1**: Indicates a perfect positive linear relationship (as one variable increases, the other increases proportionally).
*   **0**: Indicates no linear relationship.
*   **-1**: Indicates a perfect negative linear relationship (as one variable increases, the other decreases proportionally).

A **Heatmap** is a graphical representation of data where the individual values contained in a matrix are represented as colors. When applied to a correlation matrix, it visually displays the strength and direction of correlations between pairs of variables. Darker or lighter colors usually indicate stronger correlations (either positive or negative), while colors near the middle of the spectrum (e.g., white or light gray) indicate weaker or no correlations.

In the generated heatmap:

*   Each square represents the correlation coefficient between the variable on its row and the variable on its column.
*   The diagonal elements are always 1, as a variable is perfectly correlated with itself.
*   The `annot=True` parameter displays the correlation values on the heatmap.
*   `fmt='.4f'` formats these numbers to four decimal places.

From the heatmap, you can quickly identify which pairs of variables have strong positive or negative relationships, helping in feature selection and understanding the underlying structure of your data.

```python
import seaborn as sns

sns.heatmap(
    df[['Hours_Studied','Attendance','Marks']].corr(),
    annot=True,fmt='.4f'
)
print(df[['Hours_Studied','Attendance','Marks']].corr())
plt.title('Correlation Matrix')
plt.show()
```

## **Pie Chart**

A **Pie Chart** is a circular statistical graphic, which is divided into slices to illustrate numerical proportion. In a pie chart, the arc length of each slice (and consequently its central angle and area) is proportional to the quantity it represents.

In the generated pie chart:

*   Each slice represents a 'Student' and their corresponding 'Marks'.
*   The size of each slice is proportional to the student's 'Marks' relative to the total 'Marks' of all students.
*   The `autopct='%1.1f%%'` parameter formats the percentage value shown on each slice, displaying one decimal place.
*   Pie charts are best used when trying to compare parts of a whole, especially when the number of categories is small (typically less than 6-8 categories). It quickly shows the dominance of certain categories.

```python
plt.pie(
    df['Marks'],
    labels=df['Student'],
    autopct='%1.1f%%'
)

plt.title("Contribution of Students to Total Marks")
plt.show()
```

## **Bubble Chart**

A **Bubble Chart** is a variation of a scatter plot that displays three dimensions of data. Two dimensions are represented by the x and y axes, and the third dimension is represented by the size of the bubbles (markers). In some cases, a fourth dimension can be represented by the color of the bubbles.

In the generated Bubble Chart:

*   **X-axis (`Hours_Studied`)**: Represents the number of hours a student has studied.
*   **Y-axis (`Marks`)**: Represents the marks obtained by the student.
*   **Bubble Size (`s=df['Attendance']*5`)**: The size of each bubble is determined by the `Attendance` of the student, multiplied by a factor of 5 to make the differences more visually apparent. This allows you to see the impact of attendance alongside hours studied and marks.
*   **Labels (`plt.text(...)`)**: The name of each student (`df['Student']`) is added next to their respective bubble, making it easy to identify individual data points.

This chart helps to visualize relationships and patterns among three variables simultaneously. For instance, you can observe if higher hours studied and attendance generally lead to higher marks, and whether any specific student deviates from this trend.

```python
plt.figure(figsize=(10,6))
plt.scatter(
    df['Hours_Studied'],
    df['Marks'],
    s=df['Attendance']*5,
    )
for i in range(len(df)):
  plt.text(
      df['Hours_Studied'][i],
      df['Marks'][i],
      df['Student'][i]
  )
plt.xlabel("Hours Studied")
plt.ylabel("Marks")
plt.title("Bubble Chart")
plt.show()
```

## **Pair Plot**

A **Pair Plot** (also known as a scatter plot matrix) is a grid of scatter plots between all possible pairs of variables in a dataset, along with histograms or kernel density estimates for each variable on the diagonal. It is a powerful tool for exploratory data analysis (EDA) to understand the relationships between multiple variables at a glance.

In the generated Pair Plot:

*   **Diagonal Plots**: These typically show the distribution of a single variable. In this case, they are histograms for 'Hours_Studied', 'Attendance', and 'Marks', giving you an idea of the frequency of values for each individual feature.
*   **Off-Diagonal Plots**: These are scatter plots showing the relationship between two different variables. For example, the plot in the first row, second column shows 'Hours_Studied' on the x-axis and 'Attendance' on the y-axis, allowing you to observe their correlation.

**Interpretation**: A pair plot helps in quickly identifying:

*   **Relationships between variables**: You can see if there's a linear, non-linear, or no apparent relationship between any two variables.
*   **Outliers**: Unusual data points that deviate significantly from the general pattern.
*   **Distributions**: The shape and spread of each variable's data.

This plot is particularly useful when you have a moderate number of variables (e.g., 2 to 10) and want to get a holistic view of the dataset's structure and inter-variable dependencies before diving into more complex modeling.

```python
import seaborn as sns

sns.pairplot(
    df[['Hours_Studied','Attendance','Marks']]
)
plt.show()
```
