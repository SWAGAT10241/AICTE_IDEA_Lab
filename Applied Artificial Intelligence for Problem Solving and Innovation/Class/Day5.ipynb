```python
import pandas as pd

data = {
    'Hours_Studied': [2, 4, 6, 8, 10, 5, 7, 3, 9, 6],
    'Attendance': [75, 80, 85, 90, 95, 82, 88, 78, 92, 86],
    'Assignments': [4, 5, 6, 7, 8, 5, 6, 4, 7, 6],
    'Previous_Score': [32, 45, 58, 72, 88, 52, 68, 40, 80, 62],
    'Marks': [35,50,65,80,95,60,75,45,88,70],
}

df = pd.DataFrame(data)

df
```

```python
from sklearn.linear_model import LinearRegression

X = df[['Hours_Studied','Attendance','Assignments','Previous_Score']]
y = df['Marks']

model = LinearRegression()
model.fit(X,y)
```

```python
new_student = pd.DataFrame({
    'Hours_Studied': [7],
    'Attendance': [90],
    'Assignments': [6],
    'Previous_Score': [70]
})

prediction = model.predict(new_student)
print('Predicted Marks:',prediction[0])
```

```python
import matplotlib.pyplot as plt

# Predict marks for the entire dataset used for training
predicted_marks = model.predict(X)

plt.figure(figsize=(8, 6))
plt.scatter(y, predicted_marks, alpha=0.7)
plt.plot([y.min(), y.max()], [y.min(), y.max()], 'r--', lw=2, label='Perfect Prediction Line')
plt.title('Actual vs. Predicted Marks')
plt.xlabel('Actual Marks')
plt.ylabel('Predicted Marks')
plt.grid(True)
plt.legend()
plt.show()
```

```python
from sklearn.metrics import r2_score

r2 = r2_score(y, predicted_marks)
print(f"R-squared (R2) Score: {r2:.2f}")
```

```python
from sklearn.model_selection import train_test_split
```

```python
import numpy as np
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

mae = mean_absolute_error(y_test, y_pred)
print("MAE = ",mae)

mse = mean_squared_error(y_test, y_pred)
print("MSE =",mse)

rmse = np.sqrt(mse)
print("RMSE =",rmse)

r2 = r2_score(y_test, y_pred)
print("R2 =",r2)
```
